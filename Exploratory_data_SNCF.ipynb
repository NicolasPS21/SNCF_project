```python
matrix_correlation = df[["duree_moyenne", "nb_train_prevu","nb_annulation","nb_train_depart_retard","prob_percent_cause_ext", "prob_percent_infra", "prob_percent_gestion_trafic","prob_percent_materiel_roulant", "prob_percent_gestion_gare", "prob_percent_voyageurs"]].corr()
```

```python
import seaborn as sns

sns.heatmap(matrix_correlation, annot = True)
```

```python
df.describe()
```

```python
mean_station_departure = df.groupby("gare_depart")[["nb_annulation","nb_train_prevu", "retard_moyen_tous_dept", "nb_train_depart_retard","nb_train_retard_sup_15"]].mean().round(2).sort_values(by=["nb_annulation", "retard_moyen_tous_dept"], ascending=False)
```

```python
mean_station_departure
```

```python
#we gonna focus on cancelled trains for the train station.

mean_station_departure["taux_annulation"] = (mean_station_departure["nb_annulation"] / mean_station_departure["nb_train_prevu"])*100
```

```python
mean_station_departure.sort_values(by = "taux_annulation", ascending= False)
```

```python
#we want to focus on late rate

mean_station_departure["taux_retard"] = (mean_station_departure["nb_train_depart_retard"] / mean_station_departure["nb_train_prevu"])*100
```

```python
mean_station_departure.sort_values(by = "taux_retard", ascending= False)
```

```python
mean_station_departure["taux_retard_15"] = (mean_station_departure["nb_train_retard_sup_15"] / mean_station_departure["nb_train_prevu"])*100
```

```python
mean_station_departure.sort_values(by = "taux_retard_15", ascending= False)
```
